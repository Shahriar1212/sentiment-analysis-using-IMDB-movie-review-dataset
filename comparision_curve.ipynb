```python
n = 10

pyplot.plot(history_lstm['accuracy'][:n], color='r')
pyplot.plot(history_bilstm['accuracy'][:n], color='b')
pyplot.plot(history_cnn['accuracy'][:n], color='y')
pyplot.plot(history_rnn['accuracy'][:n], color='k')


pyplot.title('accuracy curve')
pyplot.ylabel('accuracy')
pyplot.xlabel('epoch')
pyplot.legend(['LSTM', 'BiLSTM', 'CNN', 'RNN'], loc='lower right')
pyplot.show()



pyplot.plot(history_lstm['loss'][:n], color='r')
pyplot.plot(history_bilstm['loss'][:n], color='b')
pyplot.plot(history_cnn['loss'][:n], color='y')
pyplot.plot(history_rnn['loss'][:n], color='k')



pyplot.title('loss curve')
pyplot.ylabel('loss')
pyplot.xlabel('epoch')
pyplot.legend(['LSTM', 'BiLSTM', 'CNN', 'RNN'], loc='upper right')
pyplot.show()
```
